```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data_path = "../data/Cleaned_Student_Performance_Factors.csv"
data = pd.read_csv(data_path)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics for numeric columns
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Ensure all columns are numeric
numeric_data = data.select_dtypes(include=[float, int])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# observations :-
'''
Correlation Values: The heatmap will display correlation coefficients between pairs of variables. These values range from -1 to 1.

1 ) 1 indicates a perfect positive correlation. -1 indicates a perfect negative correlation.
0 indicates no correlation.

2) Color Intensity: The color intensity represents the strength of the correlation.
Darker colors (closer to red or blue) indicate stronger correlations (positive or negative).
Lighter colors (closer to white) indicate weaker correlations.

'''
```

```python
# Scatter plot: Hours Studied vs Exam Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=data)
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
# obervations :-
'''
1. Trend: Look for a trend in the data points.
   - If the points form an upward trend, it indicates a positive correlation between hours studied and exam scores.
   - If the points form a downward trend, it indicates a negative correlation.
   - If the points are scattered without any clear trend, it indicates no correlation.

2. Strength of Relationship: The closeness of the points to a straight line indicates the strength of the relationship.
   - Points closely clustered around a line suggest a strong relationship.
   - Points widely scattered suggest a weak relationship.

3. Outliers: Identify any points that are far away from the general trend.
   - These outliers might indicate unusual cases or errors in the data.

4. Distribution: Observe the distribution of the data points.
   - Are there more points at certain ranges of hours studied or exam scores?
   - This can give insights into common study habits and their impact on scores.

5. Clusters: Look for any clusters of points.
   - Clusters might indicate subgroups within the data, such as students who study similar hours and achieve similar scores.

'''
```

```python
# Histogram: Distribution of Exam Scores
plt.figure(figsize=(8, 6))
data['Exam_Score'].hist(bins=20, color='red', alpha=0.7)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# observations :-
'''
1) Shape of Distribution: Look at the overall shape of the histogram.
Normal Distribution: If the histogram is bell-shaped, it indicates a normal distribution.
Skewed Distribution: If the histogram is skewed to the left or right, it indicates a skewed distribution.

2) Central Tendency: Identify where the peak of the histogram is.
The peak represents the most common exam scores.
The mean and median can be inferred from the central tendency.

3) Spread: Observe the spread of the data.
A wide spread indicates high variability in exam scores.
A narrow spread indicates low variability.
'''
```

```python
# Pie Chart: Parental Education Levels
plt.figure(figsize=(8, 6))
data['Parental_Education_Level'].value_counts().plot.pie(autopct="%1.1f%%", startangle=90)
plt.title("Parental Education Levels")
plt.ylabel("")
plt.show()
```

```python
# observations :-
'''
1) Proportions: The pie chart will show the proportion of each category of parental education levels.
Each slice represents a different education level.
The size of each slice indicates the proportion of that education level in the dataset.

2)Dominant Categories: Identify which education levels are most and least common.
Larger slices indicate more common education levels.
Smaller slices indicate less common education levels.

'''
```

```python
# Box plot: Exam Score by Parental Involvement
plt.figure(figsize=(8, 6))
sns.boxplot(x='Parental_Involvement', y='Exam_Score', data=data)
plt.title("Exam Score by Parental Involvement")
plt.xlabel("Parental Involvement")
plt.ylabel("Exam Score")
plt.show()
```

```python
# observations :-
'''
1) Median: The line inside each box represents the median exam score for each level of parental involvement.
Compare the medians to see if there is a difference in central tendency across different levels of parental involvement.

2) Interquartile Range (IQR): The box represents the IQR, which is the range between the 25th percentile (Q1) and the 75th percentile (Q3).
A larger IQR indicates more variability in exam scores for that level of parental involvement.
A smaller IQR indicates less variability.

3) Whiskers: The lines extending from the box (whiskers) show the range of the data, excluding outliers.
The length of the whiskers indicates the spread of the data.

4)Outliers: Points outside the whiskers are considered outliers.
Identify any outliers to see if there are unusual exam scores for certain levels of parental involvement.

'''
```
